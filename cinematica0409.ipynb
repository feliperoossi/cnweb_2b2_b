# Formula da Velocidade Média

```
def calcVelMedia(deltaS, deltaT):
  velMedia = (deltaS / deltaT)
  return velMedia
```

```
print('Velocidade Média: ',calcVelMedia(50, 4))
```

#Fórmula da aceleração

```
def calcVelMedia(velocidade, tempo):
  velMedia = (velocidade / tempo)
  return velMedia
print("Aceleração =", calcVelMedia(15,2))
```

#Fórmulas do Movimento Uniforme (MU)

```
def calcularPosicaoMU(posicao_inicial, velocidade, tempo):
  posicao_final = posicao_inicial + (velocidade * tempo)
  return posicao_final
```

#Função horária do deslocamento

```
def calcularDeslocamento(velocidade, tempo):
  deslocamento = velocidade * tempo
  return deslocamento
print("Função horária ", calcularDeslocamento(15, 2))
```

#Fórmulas do Movimento Uniformemente Variado (MUV)

```
def calcularVelocidadeMUV(velocidade_inicial, aceleracao, tempo):

  velocidade_final = velocidade_inicial + (aceleracao * tempo)
  return velocidade_final
print("Movimento Variado ", calcularVelocidadeMUV(15, 2,2))
```

#Equação de Torricelli

```
import math

def calTorricelli(vZero, acel, desloc):
  t=2 * acel * desloc
  vFinal = t + vZero
  res = round(math.sqrt (vFinal), 2)
  return res
```

#Fórmula do movimento vertical

```
def posicao_vertical(t, v0=0, h0=0, g=9.8):
    h = h0 + (v0 * t) + (0.5 * g * (t**2))
    return h
print(f"Altura percorrida: {posicao_vertical(t=3)} metros")
```

#Equação de Torricelli para lançamento vertical

```
def torricelli_vertical(v0, delta_h, g=9.8, subida=True):
    """Calcula a velocidade final no movimento vertical."""
    # Se está subindo, a gravidade subtrai velocidade; se descendo, soma.
    if subida:
        v_quadrado = v0**2 - 2 * g * delta_h
    else:
        v_quadrado = v0**2 + 2 * g * delta_h

    # Tratamento para evitar erro de raiz de número negativo no ponto máximo
    if v_quadrado < 0:
        return 0
    return v_quadrado**0.5

# Exemplo: Velocidade de um objeto lançado a 20m/s após subir 5 metros
v_atual = torricelli_vertical(v0=20, delta_h=5)
print(f"Velocidade atual: {v_atual:.2f} m/s")
```

#Velocidade vertical em função do tempo

```
def velocidade_vertical(t, v0=0, g=9.8):
    """
    Calcula a velocidade final (v) em um instante t.
    Use g positivo para queda e negativo para subida (se v0 for positivo).
    """
    v = v0 + (g * t)
    return v

# Exemplo: Velocidade após 2 segundos em queda livre (g=9.8)
v_inst = velocidade_vertical(t=2)
print(f"A velocidade no instante t=2s é {v_inst:.2f} m/s")
```

#Posição vertical em função do tempo

```
def posicao_vertical(t, h0=0, v0=0, g=9.8):
    """
    Calcula a posição vertical final (h) no instante t.
    """
    h = h0 + (v0 * t) + (0.5 * g * (t**2))
    return h

# Exemplo: Posição após 3 segundos partindo do repouso
altura = posicao_vertical(t=3)
print(f"A posição vertical após 3s é {altura:.2f} metros")
```
