# Formula da Velocidade Média

```
def calcVelMedia(deltaS, deltaT):
  velMedia = (deltaS / deltaT)
  return velMedia
```

teste de serviço: velocidade media

```
print('Velocidade Média: ',calcVelMedia(50, 4))
```

# Fórmulas do Movimento Uniforme (MU)

```
def calcularPosicaoMU(posicao_inicial, velocidade, tempo):
  posicao_final = posicao_inicial + (velocidade * tempo)
  return posicao_final
```

## Função horária do deslocamento

```
def calcularDeslocamento(velocidade, tempo):
  deslocamento = velocidade * tempo
  return deslocamento
print("Função horária ", calcularDeslocamento(15, 2))
```

# Fórmulas do Movimento Uniformemente Variado (MUV)

```
def calcularVelocidadeMUV(velocidade_inicial, aceleracao, tempo):

  velocidade_final = velocidade_inicial + (aceleracao * tempo)
  return velocidade_final
print("Movimento Variado ", calcularVelocidadeMUV(15, 2,2))
```

## Equação de Torricelli

```
import math

def calTorricelli(vZero, acel, desloc):
  t=2 * acel * desloc
  vFinal = t + vZero
  res = round(math.sqrt (vFinal), 2)
  return res
```
